```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread("/kaggle/input/images/PXL_20230621_125429664.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread("/kaggle/input/images/PXL_20230621_125429664 midas.png")
```

```python
def normalize_img(image):
    return (image - image.min()) / (image.max()-image.min())
```

```python
# call imshow() using plt object
norm_img1 = normalize_img(img1)
norm_img2 = normalize_img(img2)
plt.imshow(norm_img1)
  
# display that image
plt.show()
```
